```python
import math
                                        # To read data from the file

path = '/home/dr-pc/Desktop/Subjects/AI/class/Assignment#01/File1'
# path = '/home/dr-pc/Desktop/Subjects/AI/class/Assignment#01/File2'
# path = '/home/dr-pc/Desktop/AI_Assignment/File3'
# path = '/home/dr-pc/Desktop/AI_Assignment/File4'
# path = '/home/dr-pc/Desktop/AI_Assignment/File5'

with open(path,'r') as file:
    text = file.read()

                                        # Initiliaze some values and storing

clause = text.strip()
charactor_list = list(clause)
charactors = 0
words = 0
sentences = 0

                                        # For calculating the charactors

for i in charactor_list:
    if i != ' ' and i != '!' and i != '.' and i != ',' and i != '-':
        charactors += 1

                                        # For calculating the words

clause = clause.replace('-',' ')
clause = clause.replace(',',' ')
list_words = clause.split()
for i in list_words:
    if i != '-':
        words += 1

                                        # For calculating the sentences

for i in clause:
    if i == '.' or i == '!' or i == '?':
        sentences += 1

                                        # Automated Readibility Index

ARI = 4.71*(charactors/words)+0.5*(words/sentences)-21.43
ARI = max(ARI,0)                        # 0 is thresold to set the minimum value in the code
float_number = ARI - int(ARI)

if 0.1 <= float_number <= 0.6:
    ARI = math.ceil(ARI)
else:
    ARI = round(ARI)

                                        # For Printing the statement

print(f'Sentences: {sentences}')
print(f'Words: {words}')
print(f'Charactors: {charactors}')
print(f'Readiblity Index: {ARI}')
```
